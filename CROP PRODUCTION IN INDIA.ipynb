# Data Visualization

## Chart 1

```python
# Assuming 'crops' is your DataFrame
new_df = crops.groupby('Crop')['Production'].sum().reset_index()
sorted_prod = new_df.sort_values(by='Production', ascending=False)

# Ensure coconut is included, and then select the top 5 crops
top_crops = sorted_prod.head(5)  # Get the top 5 crops

# Check if coconut is in the top crops; if not, include it
if 'Coconut' not in top_crops['Crop'].values:
    # If coconut is not in the top 5, add it separately
    coconut_data = sorted_prod[sorted_prod['Crop'] == 'Coconut']
    if not coconut_data.empty:
        top_crops = pd.concat([top_crops, coconut_data]).drop_duplicates().head(5)

# Prepare data for bar chart
crop_names = top_crops['Crop']
production_values = top_crops['Production']

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(crop_names, production_values, color='blue')
plt.xlabel('Crops')
plt.ylabel('Total Production')
plt.title('Crop Production Distribution')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

**Insight:** Coconut production is the highest

## Chart 2

```python
# Assuming 'crops' is your DataFrame
new_df = crops.groupby('Crop')['Production'].sum().reset_index()
sorted_prod = new_df.sort_values(by='Production', ascending=False)

# Select crops for the pie chart 
plot_pie = sorted_prod[3:10]

# Prepare data for pie chart
crop_names = plot_pie['Crop']
production_values = plot_pie['Production']

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(production_values, labels=crop_names, autopct='%1.1f%%', startangle=140)
plt.title('Crop Production Distribution (Excluding Coconut, Sugarcane, Rice)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Insight:** Production of top 8 crops excluding coconut, sugarcane, rice are given 

## Chart 3

```python
# Assuming 'crops' is your DataFrame
state_production = crops.groupby('State_Name')['Production'].sum()

sorted_states_desc = state_production.sort_values(ascending=False)

# Get the top 15 states
top_15_states_desc = sorted_states_desc.head(15)

# Create bar plot including the top state
plt.figure(figsize=(12, 6))
plt.bar(top_15_states_desc.index, top_15_states_desc.values, color='blue')
plt.xlabel('States')
plt.ylabel('Production')
plt.title('Top 15 States by Production')
plt.xticks(rotation=90)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

**Insight:** Top 15 States Production of Crops

## Chart 4

```python
# Assuming 'crops' is your DataFrame
state_production = crops.groupby('State_Name')['Production'].sum()

# Sort states by production in ascending order
sorted_states_asc = state_production.sort_values(ascending=True)

# Get the bottom 15 states
bottom_15_states_asc = sorted_states_asc.head(15)

# Create bar plot for the bottom 15 states
plt.figure(figsize=(12, 6))
plt.bar(bottom_15_states_asc.index, bottom_15_states_asc.values, color='orange')
plt.xlabel('States')
plt.ylabel('Production')
plt.title('Bottom 15 States by Production')
plt.xticks(rotation=90)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

**Insight:** 15 Lowest Crops Production in States

## Chart 4

```python
# Group and sum the production by district
district_production = crops.groupby('District_Name')['Production'].sum()

# Sort the districts in descending order
sorted_districts_desc = district_production.sort_values(ascending=False)

# Get the top 21 districts, including number 1
top_21_districts_desc = sorted_districts_desc.head(21)

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_21_districts_desc.index, top_21_districts_desc.values)
plt.xlabel('Districts')
plt.ylabel('Production')
plt.title('Top 21 Districts by Production')
plt.xticks(rotation=90)
plt.show()
```

Insight: Top 20 District Production of Crops

## Chart 5

```python
# Filter data for Andhra Pradesh
df1 = crops
kerala_data = df1[df1['State_Name'] == 'Andhra Pradesh']

# Group and sum the production by crop
crop_production = kerala_data.groupby('Crop')['Production'].sum().reset_index()

# Sort the crops by production
prominent_crops = crop_production.sort_values('Production', ascending=False)

# Get the top 6 crops, including the first
top_6_crops = prominent_crops.head(6)  # Use head(6) instead of iloc[1:7]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_6_crops['Crop'], top_6_crops['Production'])
plt.xlabel('Crop')
plt.ylabel('Production')
plt.title('Top 6 Prominent Crops in Andhra Pradesh')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Insight: Top 6 Prominent Crops in Andhra Pradesh

```python
# Filter data for Andhra Pradesh
df1 = crops
kerala_data = df1[df1['State_Name'] == 'Uttar Pradesh']

# Group and sum the production by crop
crop_production = kerala_data.groupby('Crop')['Production'].sum().reset_index()

# Sort the crops by production
prominent_crops = crop_production.sort_values('Production', ascending=False)

# Get the top 6 crops, including the first
top_6_crops = prominent_crops.head(6)  # Use head(6) instead of iloc[1:7]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_6_crops['Crop'], top_6_crops['Production'])
plt.xlabel('Crop')
plt.ylabel('Production')
plt.title('Top 6 Prominent Crops in Uttar Pradesh')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Insight:** Top 6 Prominent Crops in Uttar Pradesh
